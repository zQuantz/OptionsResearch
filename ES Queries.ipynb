```python
from elasticsearch import Elasticsearch
```

```python
es = Elasticsearch(["34.95.46.239:8607"])
```

```python
q = es.search(index="news", body = {
    "query" : {
        "bool" : {
            "must" : [
                {
                    "match" : {
                        "tickers" : "TSLA"
                    }
                }
            ],
            "filter" : 
            [
                {
                    "range" : {
                        "oscrap_timestamp" : {
                            "gte" : "2020-07-13 14:30:00",
                            "lte" : "2020-07-13 14:35:00"
                        }
                    }
                }
            ]
        },
    }
})
```
