```python
import pandas_market_calendars as mcal
import pandas as pd
```

```python
options = pd.read_csv("data/options.csv")
options = options[
    [
        col
        for col in options.columns
        if "tdays_to" not in col
    ]
]
```

```python
options.head()
```

```python
nyse = mcal.get_calendar('NYSE')
min_date = options.date_current.min()
max_date = options.expiration_date.max()
```

```python
schedule = nyse.schedule(start_date=min_date, end_date=max_date)
trading_days = mcal.date_range(schedule, frequency="1D").tolist()
trading_days = [str(day)[:10] for day in trading_days]
```

```python
def get_trading_days(x):
    try:
        i = trading_days.index(x[0])
        j = trading_days.index(x[1])
        return j - i
    except Exception as e:
        print(e)
        return None

pairs = options[['date_current', 'expiration_date']].drop_duplicates(ignore_index=True)
pairs['tdays_to_expiry'] = pairs.apply(get_trading_days, axis=1)
pairs = pairs.dropna()
```

```python
options = options.merge(pairs, on=['date_current', 'expiration_date'], how='inner')
options.to_csv("data/options.csv", index=False)
```
