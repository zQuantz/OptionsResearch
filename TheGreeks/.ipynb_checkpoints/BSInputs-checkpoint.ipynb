```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
import sys, os
```

### Distributions

```python
x1 = np.random.normal(loc=5, scale=2, size=1_000_000)
x2 = np.random.normal(loc=-10, scale=8, size=1_000_000)
x3 = np.random.normal(loc=10, scale=6, size=1_000_000)
```

```python
f, ax = plt.subplots(1, 2, figsize=(15, 4))

ax[0].set_title("Raw")
sns.kdeplot(x1, color="blue", ax=ax[0], label="$\mu=5$ $\sigma=2$")
sns.kdeplot(x2, color="orange", ax=ax[0], label="$\mu={-3}$ $\sigma=8$")
sns.kdeplot(x3, color="red", ax=ax[0], label="$\mu=10$ $\sigma=6$")

ax[1].set_title("Transformed $(x - \mu) / \sigma$")
sns.kdeplot((x1 - x1.mean()) / x1.std(), color="blue", ax=ax[1], label="$\mu=0$ $\sigma=1$")
sns.kdeplot((x2 - x2.mean()) / x2.std(), color="orange", ax=ax[1], label="$\mu=0$ $\sigma=1$")
sns.kdeplot((x3 - x3.mean()) / x3.std(), color="red", ax=ax[1], label="$\mu=0$ $\sigma=1$")

plt.savefig("plots/raw_transformed.png")
```

### Ln Factors

```python
S = np.arange(10, 500, 1)
K = 200

plt.figure(figsize=(14, 9))
sns.scatterplot(S, np.log(S / K))
plt.xlabel("Stock Price", size=18)
plt.ylabel("$ln(S/K)$          ", rotation=0, size=18)
plt.xticks(size=18)
plt.yticks(size=18)
plt.savefig("plots/moneyness.png")
```

```python
S = np.arange(10, 500, 1)
K = 200

plt.figure(figsize=(14, 9))
sns.scatterplot(S, np.log(S / K))
plt.xlabel("Stock Price", size=18)
plt.ylabel("$ln(S/K)$          ", rotation=0, size=18)
plt.axhline(y=0, color="black")
plt.axvline(x=K, color="black")
plt.xticks(size=18)
plt.yticks(size=18)
plt.savefig("plots/moneyness2.png")
```

### d1 and d2

```python
r = 0.001
sigma = 0.3
T = 0.25

d1 = np.log(S / K) + (r - sigma*sigma * 0.5) * np.sqrt(T)
d1 /= sigma * np.sqrt(T)

d2 = d1 - sigma * np.sqrt(T)
```
